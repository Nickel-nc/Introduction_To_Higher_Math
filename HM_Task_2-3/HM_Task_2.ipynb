### 1. Задание
Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20)
- 1)	Найдите их сумму. (на листочке)

`Ответ:`

$r_1=\left(\begin{array}{crl}20\\20\\20\end{array}\right)\;; \quad r_2=\left(\begin{array}{crl}0\\0\\-20\end{array}\right)\;; \\
r_1 + r_2 = \left(\begin{array}{crl}20+0\\20+0\\20-20\end{array}\right) = \left(\begin{array}{crl}20\\20\\0\end{array}\right)$

- 2)	Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
r1 = np.array([20,20,20])
r2 = np.array([0,0,-20])
r3 = r1+r2

def calculate_vector_length(coordinates):
    length = (coordinates[0]**2 + coordinates[1]**2 + coordinates[2]**2)**(0.5)
    return length

print(f'Длина вектора: {calculate_vector_length(r3)}')
```

Отображение угла искажается размерностью координатной сетки по осям `x` и `y`

Для корректного отображения можно настроить график с помощью метода `plt.axes().set_aspect(1)`

```python
x = np.linspace(-5,5,21)
y = 3 *x+1
y2 = (-1/3)*x+1

plt.plot(x,y)
plt.plot(x,y2)
plt.axes().set_aspect(1)
plt.show()
```

### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
- 1	окружности,
- 2	эллипса,
- 3	гиперболы.

```python
t=np.linspace(0,2*np.pi,100)
k=1
h=2
r=2
plt.axes().set_aspect(1)
plt.plot(np.cos(t)*r+k,np.sin(t)*r+h,'r--')
plt.grid(True)
plt.show()
```

### 4. Задание (на листочке)
1) Пусть задана плоскость $Ax + Bx + Cz + D = 0$ 

Напишите уравнение плоскости, параллельной данной и проходящей через начало координат.


`Ответ:`

- Для того, чтобы плоскости были параллельны, должно выполняться условие равенства коэффициентов `A, B, C`, задающих вектор нормали к плоскости
- Плоскость будет проходить через начало координат (0.0.0) если приравнять коэффициент `D` к нулю 

2)  Пусть задана плоскость: $A_1x + B_1y + C_1z + D_1 = 0$
и прямая: 

 $\frac{x-x_1}{x_2-x_1}=\frac{y-y_1}{y_2-y_1} = \frac{z-z_1}{z_2-z_1}$
 
Как узнать, принадлежит прямая плоскости или нет?


`Ответ:`

Прямая принадлежит плоскости, если две точки прямой принадлежат этой плоскости или, если она имеет общую точку с плоскостью и параллельна прямой, лежащей в этой плоскости

Таким образом будет выполняться равенство

$A(x-x_1)+B(y-y_1)+C(z-z_1)=0 \\
A(x-x_2)+B(y-y_2)+C(z-z_2)=0$

- 1)	Нарисуйте трехмерный график двух параллельных плоскостей.

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
A = 2
B = 3
C = 0
D = 150
X, Y = np.meshgrid(X,Y)
Z1 = A*X + B*Y + D
Z2 = A*X + B*Y
ax.plot_wireframe(X, Y, Z1)
ax.plot_wireframe(X, Y, Z2)
ax.scatter(0, 0, 0, 'z', 50, 'red')
show()
```
